````python
# Verificar si TensorFlow detecta la GPU
print("Dispositivos disponibles:", tf.config.list_physical_devices('GPU'))

# Ruta de la carpeta con las imágenes
ruta_carpeta = r'C:\Users\User\Documents\AI-and-data-analysis-subject-and-projects\Exercise CNN\dtaset\images'

# Obtener lista de todas las imágenes en la carpeta
imagenes = [f for f in os.listdir(ruta_carpeta) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
imagenes = natsorted(imagenes)


# Función para cargar y convertir a escala de grises usando TensorFlow
def convertir_a_grises_tf(imagen):
    # Cargar la imagen
    imagen = tf.io.read_file(imagen)
    imagen = tf.image.decode_jpeg(imagen, channels=3)  # Cargar como RGB
    imagen_gris = tf.image.rgb_to_grayscale(imagen)  # Convertir a escala de grises
    return imagen_gris

# Lista para almacenar las imágenes en escala de grises
imagenes_grises = []

# Procesar las imágenes usando TensorFlow
for imagen_nombre in imagenes:
    ruta_imagen = os.path.join(ruta_carpeta, imagen_nombre)
    
    # Convertir a escala de grises usando TensorFlow
    imagen_gris = convertir_a_grises_tf(ruta_imagen)
    
    # Ejecutar en sesión para obtener el valor de la imagen (en caso de estar en un gráfico de operaciones)
    imagen_gris = imagen_gris.numpy()  # Convierte el tensor de vuelta a un array NumPy
    imagenes_grises.append(imagen_gris)

print(f"Total de imágenes procesadas: {len(imagenes_grises)}")


````

````python
# Tamaño deseado para todas las imágenes (alto, ancho)
nuevo_tamano = (256, 256)  # por ejemplo

# Redimensionar todas las imágenes en la lista
imagenes_redimensionadas = [
    cv2.resize(img.squeeze(), nuevo_tamano, interpolation=cv2.INTER_AREA)
    for img in imagenes_grises
]
for i in range(5):
    # Visualizar el primer elemento de la lista usando Matplotlib
    plt.imshow(imagenes_redimensionadas[i].squeeze(), cmap='gray')  # .squeeze() elimina cualquier dimensión extra
    plt.axis('off')  # No mostrar los ejes
    plt.show()

plt.imshow(imagenes_redimensionadas[7390].squeeze(), cmap='gray')  # .squeeze() elimina cualquier dimensión extra
plt.axis('off')  # No mostrar los ejes
plt.show()
````

# Procedemos a hacer nuestras cNN custom, en este caso con 3 modelos

````python
# Suponiendo que imagenes_redimensionadas es una lista de imágenes tf.Tensor escala de grises (256, 256)
# Paso 1: Asegurarse de que todas tengan canal (1) explícito
imagenes_con_canal = [tf.expand_dims(img, axis=-1) for img in imagenes_redimensionadas]

# Paso 2: Convertir a un solo tensor de forma (N, 256, 256, 1)
imagenes_final = tf.stack(imagenes_con_canal)

print(imagenes_final.shape)  # (num_imagenes, 256, 256, 1)
````

````python
X_1 = imagenes_con_canal
X_2 = imagenes_final
````

````python
X_1 = np.array(X_1).astype(float) / 255.0
X_2 = np.array(X_2).astype(float) / 255.0
````

````python
y = cat_dog_labels_codified
````

````python
y=np.array(y)
````

## El primer modelo (M1) tendra solo una capa de entrada que recibe los pixeles, dos capas densas de 150 neuronas y una capa de salIda que sea se una neurona con salida tipo sigmoid que devolvera un 0 o un 1 en funcion de si es PERRO (1) O GATO(0)

````python
modeloDenso = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape = (256, 256, 1)),
    tf.keras.layers.Dense(150, activation = 'relu'),
    tf.keras.layers.Dense(150, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])
````

## Para este modelo si que probaremos de usar kernels para hacer la red convolucional

````python
modeloCNN = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation = 'relu', input_shape = (256, 256, 1)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(100, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])
````

## Compilamos los modelos

````python
modeloDenso.compile(optimizer= 'adam',
                   loss= 'binary_crossentropy',
                   metrics=['accuracy'])
````

````python
modeloCNN.compile(optimizer= 'adam',
                   loss= 'binary_crossentropy',
                   metrics=['accuracy'])
````

````python
# Ruta donde guardar los logs
log_dir = os.path.join("logs", "denso")
tensorboardDenso = TensorBoard(log_dir = log_dir)
modeloDenso.fit(X_1, y, batch_size = 32,
               validation_split = 0.15,
               epochs = 100,
               callbacks = [tensorboardDenso])

# Ruta donde guardar los logs
log_dir = os.path.join("logs", "cnn")
tensorboardCNN = TensorBoard(log_dir = log_dir)
modeloCNN.fit(X_1, y, batch_size = 32,
               validation_split = 0.15,
               epochs = 100,
               callbacks = [tensorboardDenso])
````

Guardamos los modelos

````python
modeloDenso.save("modeloDenso.keras")
modeloCNN.save("modeloCNN.keras")
````
