## en primer lugar importamos los datos de la universidad de california con la lireria ucimlrepo 

```python
import pandas as pd
from ucimlrepo import fetch_ucirepo 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# fetch dataset 
wine_quality = fetch_ucirepo(id=186) 
# data (as pandas dataframes) 
X = wine_quality.data.features 
y = wine_quality.data.targets 
  
df=pd.concat([X,y],axis=1)
```

# Data exploration

en primer lugar, observaremos la distribucion de las calidades de vino ya que la parte donde comprobamos si hay valores nulos, si hay que limpoaar los datos se ha hecho anteriormente y por suerte los datos aportados son ya bastante limpios

```python
X_df = pd.DataFrame(X, columns=[
    'fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar', 
    'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density',
    'pH', 'sulphates', 'alcohol'
])

# 1. Visualización de la distribución de la variable objetivo (quality)
plt.figure(figsize=(8, 6))
sns.histplot(y, bins=10, kde=True, color='teal')
plt.title('Distribución de la Calidad del Vino')
plt.xlabel('Calidad del Vino')
plt.ylabel('Frecuencia')
plt.show()
```

como se puede observar, encontramos que la mayoria de los vinos tienen un 6 de calidad ya demas las calidades siguen una distribucion bastante normal

```python
# 2. Análisis de la correlación entre las características
plt.figure(figsize=(10, 8))
correlation_matrix = X_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación entre las Características')
plt.show()
```

como bien se puede observar en la matriz de correlacion, vemos relaciones que son obvias. Por ejemplo es normal que haya poca relacion entre variables como la acidez volatil y la cantidad de alcohol, ya que los compuestos quimicos relacionados con estas variables no son susceptibles de reaccionar. En cambio cuando vemos relaciones entre el totald e dioxido de sulfuro y el dioxido de sulfuro libre hay mucha correlacion ya que en ambos casos hablamos de el mismo compuesto solo que en distintas cantidades

```python
# 4. Distribución de las características más relevantes
fig, axes = plt.subplots(3, 4, figsize=(16, 12))  # 3 filas, 4 columnas
axes = axes.flatten()

for i, column in enumerate(X_df.columns):
    sns.histplot(X_df[column], bins=20, kde=True, ax=axes[i], color='dodgerblue')
    axes[i].set_title(f'Distribución de {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

de esta visualizacion podemos ver algo curioso. Por ejemplo: variables como la fixed_acidity y la volatile_acidity siguen distribuciones tipo chisquared, mientras que variables como el pH siguen distribuiciones normales. De esta manera podemos preever mas o menos que un vino puede obtener valores de pH cercanos a 3.25 mientras que otras variables como por ejemplo el porcentaje de alcohol siempre tiende a ser del 10%.

```python
# 5. Visualización de la relación entre la calidad y el alcohol (variable comúnmente importante)
plt.figure(figsize=(8, 6))
sns.boxplot(x=y.values.ravel(), y=X_df['alcohol'], palette="coolwarm")
plt.title('Relación entre Calidad y Alcohol')
plt.xlabel('Calidad del Vino')
plt.ylabel('Alcohol')
plt.show()

# 6. Visualización de la relación entre la calidad y el pH
plt.figure(figsize=(8, 6))
sns.boxplot(x=y.values.ravel(), y=X_df['pH'], palette="muted")
plt.title('Relación entre Calidad y pH')
plt.xlabel('Calidad del Vino')
plt.ylabel('pH')
plt.show()
```

Finalmente, con esta ultima representacion, si fueramos una empresa consultora de vinos, y un cliente nos viene a preguntar que como puede hacer el mejor vino le recomendariamos lo siguiente: Los vinos con mayores puntuaciones son aquellos con un 13% de concentracion de alcohol y aquellos cuyo pH es de 3.4, asi que un vino acido y ademas "peleón" puede ser una gran idea para comercializar. Sin embargo, como hay modelos de ML que analizaran mas caracteristicas y mejor que nosotros, entrenaremos un modelo tipo Gradient Boost para que aprenda a predecir la calidad del vino por nosotros

## Modelo de regresion gradient boosting

```python
# 1. Importar las librerías necesarias
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# 2. Dividir los datos en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y.values.ravel(), test_size=0.2, random_state=42)

# 3. Definir el modelo de Gradient Boosting
gb_model = GradientBoostingRegressor(random_state=42)

# 4. Definir el diccionario de parámetros para GridSearchCV
param_grid = {
    'n_estimators': [250, 300],  # Número de árboles
    'learning_rate': [0.01, 0.05, 0.1, 0.2],  # Tasa de aprendizaje
    'max_depth': [3, 4, 5, 6]  # Profundidad máxima de los árboles
}

# 5. Configurar GridSearchCV
grid_search = GridSearchCV(estimator=gb_model, param_grid=param_grid, 
                           cv=5, scoring='neg_mean_squared_error', n_jobs=-1, verbose=2)

# 6. Ajustar GridSearchCV a los datos de entrenamiento
grid_search.fit(X_train, y_train)

# 7. Ver los mejores parámetros encontrados
print("Mejores parámetros encontrados: ", grid_search.best_params_)

# 8. Obtener el mejor modelo (con los mejores parámetros)
best_model = grid_search.best_estimator_

# 9. Hacer predicciones con el mejor modelo
y_pred = best_model.predict(X_test)

# 10. Evaluar el rendimiento del modelo
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

r2 = r2_score(y_test, y_pred)
print(f"R² Score: {r2:.2f}")
```

obtenemos un r^2 de 0.46, no es de los mejores, random forest conseguia mejor resultado sin embargo teniendo en cuenta que este valor va entre -1 y 1, no es un mal coeficiente de correlacion

```python
# 12. Graficar la comparación entre los valores reales y las predicciones
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')  # Línea de igualdad
plt.title('Comparación entre Valores Reales y Predicciones')
plt.xlabel('Valores Reales (Quality)')
plt.ylabel('Predicciones del Modelo')
plt.show()

# 13. (Opcional) Visualizar la importancia de las características
feature_importances = best_model.feature_importances_
plt.figure(figsize=(10, 6))
plt.barh(X_df.columns, feature_importances, color='lightblue')
plt.title('Importancia de las Características')
plt.xlabel('Importancia')
plt.ylabel('Características')
plt.show()
```

finalmente, gracias a nuestro modelo, podriamos decirle al cliente que sin lugar a dudas, la caracteristica mas impoertante del vino es l a cantidad de alcohol y la acidez volatil.

# Conclusiones finales

Las características más importantes que tu modelo de **Gradient Boosting** ha identificado, como **el alcohol del vino** y **su acidez volátil**, están fuertemente relacionadas con las propiedades organolépticas y químicas que afectan tanto el sabor como la calidad del vino. Vamos a explorar por qué estas dos características pueden ser tan influyentes para determinar la calidad del vino.

### 1. **Alcohol del vino:**
El contenido de **alcohol** en el vino es uno de los factores más importantes que influye en su **sabor, cuerpo y equilibrio**. Aquí te explico por qué:

- **Sabor y cuerpo:** El alcohol es responsable de proporcionar una sensación de cuerpo al vino. Un vino con un mayor contenido de alcohol suele sentirse más **caliente** en boca, con un sabor más robusto y con mayor persistencia.
- **Equilibrio:** La cantidad de alcohol tiene un impacto directo en el equilibrio del vino. Si el contenido de alcohol es demasiado alto o demasiado bajo en relación con otros compuestos, como los azúcares y los ácidos, el vino puede parecer desequilibrado, afectando su calidad.
- **Madurez y fermentación:** Los vinos con mayor concentración de alcohol suelen ser aquellos que han pasado por una fermentación más completa y tienen una mayor **madurez**. Esto suele correlacionarse con vinos de mayor calidad.

En resumen, el alcohol es un componente esencial porque afecta la **estructura** y la **percepción sensorial** del vino, siendo crucial para la calidad percibida.

### 2. **Acidez volátil:**
La **acidez volátil**, que incluye compuestos como el **ácido acético**, es una medida de la cantidad de ácidos volátiles presentes en el vino. Estos son los ácidos que se **evaporan rápidamente** y tienen un impacto significativo en las características sensoriales del vino.

- **Sabor:** El ácido acético es un componente principal de lo que se conoce como el **vinagre**, y en altas concentraciones, la acidez volátil puede **afectar negativamente** el sabor del vino, dándole un carácter desagradable o "rancio". Sin embargo, en niveles bajos, puede aportar una **sensación de frescura** al vino.
- **Fermentación y calidad:** La acidez volátil aumenta generalmente durante la **fermentación** y es un subproducto de la actividad microbiana. La presencia de una cantidad excesiva de acidez volátil puede ser un indicio de **defectos de vinificación** o de **contaminación** en el proceso de producción, lo que a su vez disminuye la calidad del vino.
- **Estabilidad del vino:** Un vino con una acidez volátil bien equilibrada puede tener una mayor **estabilidad** y una mayor capacidad de **envejecimiento**.

Por lo tanto, la acidez volátil es un **indicador clave de la calidad** en términos de posibles defectos, pero también de la frescura y equilibrio en el vino. Un nivel muy alto de acidez volátil generalmente indica un problema, mientras que un nivel moderado puede ser una característica deseable dependiendo del tipo de vino.

### Relación con el modelo:
Tu modelo de Gradient Boosting ha identificado que estas dos características son las más importantes para predecir la **calidad del vino**. Esto tiene sentido porque ambas características afectan de manera significativa las **sensaciones de sabor** y **aroma** que los catadores perciben, así como la estabilidad y complejidad del vino.

### Conclusión:
- **Alcohol**: Influye en el **cuerpo** y **equilibrio** del vino.
- **Acidez volátil**: Relacionada con la **frescura** y la **posibilidad de defectos** o **malas prácticas de vinificación**.

Ambas características son cruciales para la calidad del vino porque afectan las **propiedades sensoriales** y la **estabilidad** del vino a largo plazo. ¡Es por eso que tu modelo de Gradient Boosting las considera tan importantes!
