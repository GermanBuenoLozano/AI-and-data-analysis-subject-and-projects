```python
import pandas as pd
from ucimlrepo import fetch_ucirepo 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# fetch dataset 
wine_quality = fetch_ucirepo(id=186) 
# data (as pandas dataframes) 
X = wine_quality.data.features 
y = wine_quality.data.targets 
  
df=pd.concat([X,y],axis=1)
```

```python
X_df = pd.DataFrame(X, columns=[
    'fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar', 
    'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density',
    'pH', 'sulphates', 'alcohol'
])

# 1. VisualizaciÃ³n de la distribuciÃ³n de la variable objetivo (quality)
plt.figure(figsize=(8, 6))
sns.histplot(y, bins=10, kde=True, color='teal')
plt.title('DistribuciÃ³n de la Calidad del Vino')
plt.xlabel('Calidad del Vino')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# 2. AnÃ¡lisis de la correlaciÃ³n entre las caracterÃ­sticas
plt.figure(figsize=(10, 8))
correlation_matrix = X_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de CorrelaciÃ³n entre las CaracterÃ­sticas')
plt.show()
```

como bien se puede observar en la matriz de correlacion, vemos relaciones que son obvias. En cambio cuando vemos relaciones entre el totald e dioxido de sulfuro y el dioxido de sulfuro libre hay mucha correlacion ya que en ambos casos hablamos de el mismo compuesto solo que en distintas cantidades

Sin embargo, como hay modelos de ML que analizaran mas caracteristicas y mejor que nosotros, entrenaremos un modelo tipo Gradient Boost para que aprenda a predecir la calidad del vino por nosotros

## Modelo de regresion gradient boosting

```python
# 1. Importar las librerÃ­as necesarias
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# 2. Dividir los datos en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y.values.ravel(), test_size=0.2, random_state=42)

# 3. Definir el modelo de Gradient Boosting
gb_model = GradientBoostingRegressor(random_state=42)

# 4. Definir el diccionario de parÃ¡metros para GridSearchCV
param_grid = {
    'n_estimators': [250, 300],  # NÃºmero de Ã¡rboles
    'learning_rate': [0.01, 0.05, 0.1, 0.2],  # Tasa de aprendizaje
    'max_depth': [3, 4, 5, 6]  # Profundidad mÃ¡xima de los Ã¡rboles
}

# 5. Configurar GridSearchCV
grid_search = GridSearchCV(estimator=gb_model, param_grid=param_grid, 
                           cv=5, scoring='neg_mean_squared_error', n_jobs=-1, verbose=2)

# 6. Ajustar GridSearchCV a los datos de entrenamiento
grid_search.fit(X_train, y_train)

# 7. Ver los mejores parÃ¡metros encontrados
print("Mejores parÃ¡metros encontrados: ", grid_search.best_params_)

# 8. Obtener el mejor modelo (con los mejores parÃ¡metros)
best_model = grid_search.best_estimator_

# 9. Hacer predicciones con el mejor modelo
y_pred = best_model.predict(X_test)

# 10. Evaluar el rendimiento del modelo
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

r2 = r2_score(y_test, y_pred)
print(f"RÂ² Score: {r2:.2f}")
```

obtenemos un r^2 de 0.46, no es de los mejores, random forest conseguia mejor resultado sin embargo teniendo en cuenta que este valor va entre -1 y 1, no es un mal coeficiente de correlacion

```python
# 12. Graficar la comparaciÃ³n entre los valores reales y las predicciones
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')  # LÃ­nea de igualdad
plt.title('ComparaciÃ³n entre Valores Reales y Predicciones')
plt.xlabel('Valores Reales (Quality)')
plt.ylabel('Predicciones del Modelo')
plt.show()

# 13. (Opcional) Visualizar la importancia de las caracterÃ­sticas
feature_importances = best_model.feature_importances_
plt.figure(figsize=(10, 6))
plt.barh(X_df.columns, feature_importances, color='lightblue')
plt.title('Importancia de las CaracterÃ­sticas')
plt.xlabel('Importancia')
plt.ylabel('CaracterÃ­sticas')
plt.show()
```

finalmente, gracias a nuestro modelo, podriamos decirle al cliente que sin lugar a dudas, la caracteristica mas impoertante del vino es l a cantidad de alcohol y la acidez volatil.

Las caracterÃ­sticas mÃ¡s importantes que tu modelo de **Gradient Boosting** ha identificado, como **el alcohol del vino** y **su acidez volÃ¡til**, estÃ¡n fuertemente relacionadas con las propiedades organolÃ©pticas y quÃ­micas que afectan tanto el sabor como la calidad del vino. Vamos a explorar por quÃ© estas dos caracterÃ­sticas pueden ser tan influyentes para determinar la calidad del vino.

### 1. **Alcohol del vino:**
El contenido de **alcohol** en el vino es uno de los factores mÃ¡s importantes que influye en su **sabor, cuerpo y equilibrio**. AquÃ­ te explico por quÃ©:

- **Sabor y cuerpo:** El alcohol es responsable de proporcionar una sensaciÃ³n de cuerpo al vino. Un vino con un mayor contenido de alcohol suele sentirse mÃ¡s **caliente** en boca, con un sabor mÃ¡s robusto y con mayor persistencia.
- **Equilibrio:** La cantidad de alcohol tiene un impacto directo en el equilibrio del vino. Si el contenido de alcohol es demasiado alto o demasiado bajo en relaciÃ³n con otros compuestos, como los azÃºcares y los Ã¡cidos, el vino puede parecer desequilibrado, afectando su calidad.
- **Madurez y fermentaciÃ³n:** Los vinos con mayor concentraciÃ³n de alcohol suelen ser aquellos que han pasado por una fermentaciÃ³n mÃ¡s completa y tienen una mayor **madurez**. Esto suele correlacionarse con vinos de mayor calidad.

En resumen, el alcohol es un componente esencial porque afecta la **estructura** y la **percepciÃ³n sensorial** del vino, siendo crucial para la calidad percibida.

### 2. **Acidez volÃ¡til:**
La **acidez volÃ¡til**, que incluye compuestos como el **Ã¡cido acÃ©tico**, es una medida de la cantidad de Ã¡cidos volÃ¡tiles presentes en el vino. Estos son los Ã¡cidos que se **evaporan rÃ¡pidamente** y tienen un impacto significativo en las caracterÃ­sticas sensoriales del vino.

- **Sabor:** El Ã¡cido acÃ©tico es un componente principal de lo que se conoce como el **vinagre**, y en altas concentraciones, la acidez volÃ¡til puede **afectar negativamente** el sabor del vino, dÃ¡ndole un carÃ¡cter desagradable o "rancio". Sin embargo, en niveles bajos, puede aportar una **sensaciÃ³n de frescura** al vino.
- **FermentaciÃ³n y calidad:** La acidez volÃ¡til aumenta generalmente durante la **fermentaciÃ³n** y es un subproducto de la actividad microbiana. La presencia de una cantidad excesiva de acidez volÃ¡til puede ser un indicio de **defectos de vinificaciÃ³n** o de **contaminaciÃ³n** en el proceso de producciÃ³n, lo que a su vez disminuye la calidad del vino.
- **Estabilidad del vino:** Un vino con una acidez volÃ¡til bien equilibrada puede tener una mayor **estabilidad** y una mayor capacidad de **envejecimiento**.

Por lo tanto, la acidez volÃ¡til es un **indicador clave de la calidad** en tÃ©rminos de posibles defectos, pero tambiÃ©n de la frescura y equilibrio en el vino. Un nivel muy alto de acidez volÃ¡til generalmente indica un problema, mientras que un nivel moderado puede ser una caracterÃ­stica deseable dependiendo del tipo de vino.
